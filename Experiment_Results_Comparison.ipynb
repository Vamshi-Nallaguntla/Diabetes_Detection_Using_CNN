```python
import pandas as pd
import matplotlib.pyplot as plt

# Define results for each split
results = {
    "Experiment": ["60/40 Split", "70/30 Split", "80/20 Split", "5-Fold Cross Validation"],
    "Accuracy (%)": [65.58, 63.20, 68.18, 70.05],
    "Precision (%)": [47.96, 47.13, 55.56, 58.69],
    "Recall (%)": [46.08, 51.25, 54.55, 35.00],
    "F1 Score (%)": [47.00, 49.10, 55.05, 39.23],
    "AUC (%)": [71.59, 70.40, 70.56, 71.53]
}

# Convert to DataFrame
results_df = pd.DataFrame(results)

# Display the DataFrame
print("Experimental Results Comparison:")
print(results_df)

# Generate bar plots for each metric
metrics = ["Accuracy (%)", "Precision (%)", "Recall (%)", "F1 Score (%)", "AUC (%)"]
for metric in metrics:
    plt.figure(figsize=(8, 5))
    plt.bar(results_df["Experiment"], results_df[metric])
    plt.title(f"{metric} Comparison")
    plt.ylabel(metric)
    plt.xlabel("Experiment")
    plt.xticks(rotation=45, ha='right')
    for i, v in enumerate(results_df[metric]):
        plt.text(i, v + 1, f"{v:.2f}%", ha='center')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define results for each split
results = {
    "Experiment": ["60/40 Split", "70/30 Split", "80/20 Split", "5-Fold Cross Validation"],
    "Accuracy (%)": [65.58, 63.20, 68.18, 70.05],
    "Precision (%)": [47.96, 47.13, 55.56, 58.69],
    "Recall (%)": [46.08, 51.25, 54.55, 35.00],
    "F1 Score (%)": [47.00, 49.10, 55.05, 39.23],
    "AUC (%)": [71.59, 70.40, 70.56, 71.53]
}

# Convert to DataFrame
results_df = pd.DataFrame(results)

# Create a grouped bar chart
metrics = ["Accuracy (%)", "Precision (%)", "Recall (%)", "F1 Score (%)", "AUC (%)"]
x = np.arange(len(results_df["Experiment"]))  # the label locations
width = 0.15  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 8))

for i, metric in enumerate(metrics):
    ax.bar(x + i * width, results_df[metric], width, label=metric)

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Experiments')
ax.set_ylabel('Metrics (%)')
ax.set_title('Comparison of Metrics Across Experiments')
ax.set_xticks(x + width * (len(metrics) - 1) / 2)
ax.set_xticklabels(results_df["Experiment"])
ax.legend(loc="best")

# Add values on top of the bars
for i, metric in enumerate(metrics):
    for j, value in enumerate(results_df[metric]):
        ax.text(x[j] + i * width, value + 1, f"{value:.2f}%", ha='center', va='bottom', fontsize=8)

fig.tight_layout()
plt.show()
```
